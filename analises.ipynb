```python
df = df.drop_duplicates()
df.shape
```

```python
# Distribuio da varivel alvo
plt.figure(figsize=(8,5))
sns.histplot(df["CO2 Emissions(g/km)"], bins=40, kde=True)
plt.title("Distribuio das emisses de CO")
plt.xlabel("CO (g/km)")
plt.ylabel("Frequncia")
plt.show()
```

```python
num_cols = df.select_dtypes(include='number')

# Correlao entre elas
plt.figure(figsize=(10,6))
sns.heatmap(num_cols.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlao entre variveis numricas")
plt.show()
```
